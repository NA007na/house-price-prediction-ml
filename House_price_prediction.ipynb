```python
import numpy as np
from sklearn.linear_model import LinearRegression

areas=np.array([[600],[1200],[2400],[4000],[10000]])
prices=np.array([80,160,320,500,1000])
model=LinearRegression()
model.fit(areas,prices)
userinput_scalar=float(input("Enter the dimensions of the plot which you are looking for :)"))
userinput=np.array([[userinput_scalar]]) # Reshape the input to a 2D array
prediction=model.predict(userinput)
print("The estimated value is",prediction[0],"lakhs")
```
